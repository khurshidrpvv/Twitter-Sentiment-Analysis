```python
%matplotlib inline
import re
import matplotlib.pyplot as plt
```

```python
# Source has info about the type of device used for tweeting

source = []
with open("Source.txt",encoding="utf8") as f:
    for line in f:
        source.append(line)
```

```python
source[1:10]
```

```python
# Extracing only the required info e.g "Twitter for Android"
# Using regular expression

pattern = re.compile('>([a-zA-Z ]*)<')
device = []
for i in source:
    match = re.search(pattern,i)
    if match:
        device.append(match.group(1))
```

```python
len(device)
```

```python
# Saving device type and their coresponding counts i.e number of times device is used 

device_type = {}
for d in device:
    if d in device_type:
        device_type[d] +=1
    else:
        device_type[d] = 1
```

```python
# Concerting key value pair to list for plotting on pie chart

device_type = { k : v for k,v in device_type.items() if v >3000}
name = [k for k in device_type if k ]
slices = [v for v in device_type.values()]
device_type
        
```

```python
# Plotting on pie chart

%matplotlib nbagg
cols = ['b','g','r','c','m','y','w']
plt.pie(slices, labels = name ,autopct= '%1.1f%%',colors = cols,pctdistance=1.1,labeldistance=1.25)
plt.title('Tweet Device')
plt.show()
```
